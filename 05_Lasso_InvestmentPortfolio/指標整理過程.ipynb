```python
import pandas as pd
import datetime
from dateutil.relativedelta import relativedelta
```

# 讀取

```python
df=pd.read_csv('/Users/chenjessie/Desktop/經濟預測模型/美国失业率.csv')
df
```

# 更改欄位名稱

```python
df=df.drop(columns=['前值(%)'])
df.columns = ['日期','失業率']
df
```

# [1] 日期換成  xxxx-xx-xx  格式

```python
for i in range(df.shape[0]):
    y = df['日期'].iloc[i].split('年')[0]
    m = df['日期'].iloc[i].split('年')[1].split('月')[0]
    date = y+'-'+m+'-01'
    print(date)
    df['日期'].iloc[i]=date
```

# [2] 刪除重複日期，取第一筆

```python
df=df.drop_duplicates('日期','first')
df
```

# [3] 將日期設為index

```python
df['日期'] = pd.to_datetime(df['日期'])
df = df.set_index('日期') 
df
```

# [4] 創一個完整日期的pd

```python
import datetime
from dateutil.relativedelta import relativedelta
dic={}
dic['日期']=[]
in_date = '2020-04-01'
dt = datetime.datetime.strptime(in_date, "%Y-%m-%d")
for i in range(50*12+4):
    dt = datetime.datetime.strptime(in_date, "%Y-%m-%d")
    dt = (dt-relativedelta(months=i)).strftime("%Y-%m-%d")
    dic['日期'].append(dt)
    #print(dt)
    
#print(dic['Date'])
new = pd.DataFrame(dic)
new['日期'] = pd.to_datetime(new['日期'])
new = new.set_index('日期') 
new
```

# [5] Left Join

```python
res = pd.merge(new,df, on='日期',how='outer')
res
```

# [6] 用平均數填補空值

```python
print(res.isna().sum())
```

```python
res = res.replace('-', None) 
```

```python
print(res.isna().sum())
```

```python
res['信心'] = res['信心'].astype(float)
```

```python
x = res.mean(axis=0)
x
```

```python
res = res.fillna(x)
```

```python
print(res.isna().sum())
```

# 刪除2020-05-01

```python
in_date='2020-05-01'
dt = datetime.datetime.strptime(in_date, "%Y-%m-%d")
res=res.drop(index=dt)
```
